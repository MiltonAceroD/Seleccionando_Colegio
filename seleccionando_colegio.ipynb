```python
sw = False
for t in dict_archivo.items():
    
    df=pd.read_excel(t[1], skiprows=5, encoding='utf-8')
    df.columns=df.columns.str.replace('\W', '')

    df2=df[(df.Municipio=="BARRANQUILLA (ATLANTICO)") &
           (df.NombredelEstablecimiento.isin(dict_nombre_corto.keys()))][lista_campos]
    df2['Año']=t[0]
    df2['NombreCortoColegio']= df["NombredelEstablecimiento"].apply(lambda x: dict_nombre_corto.get(x))
    df2=df2[lista_campos2]
    print(df2.shape)
    if sw == False:
        df3=df2
        sw=True
    else:
        df3 = pd.concat([df3, df2], ignore_index=True, sort=False)
        #df3=df3.append(df2, ignore_index=True)
df3.ÍndiceTotal=df3.ÍndiceTotal.astype(float)
df3.ÍndicedeInglés=df3.ÍndicedeInglés.astype(float)

df3.ÍndicedeMatemática=df3.ÍndicedeMatemática.astype(float)
df3.ÍndicedeCienciasNaturales=df3.ÍndicedeCienciasNaturales.astype(float)
df3.ÍndicedeSocialesyCiudadanas=df3.ÍndicedeSocialesyCiudadanas.astype(float)
df3.ÍndicedeLecturaCrítica=df3.ÍndicedeLecturaCrítica.astype(float)


```
